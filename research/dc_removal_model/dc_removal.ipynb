#DC removal

```python
import numpy as np
from scipy.io import wavfile
from scipy.signal import butter, filtfilt

def high_pass_filter(data, sample_rate, cutoff_freq=100):
    """
    Apply a high-pass filter to remove frequencies below the cutoff_freq.

    Parameters:
    - data: The input audio signal (numpy array).
    - sample_rate: The sample rate of the audio signal.
    - cutoff_freq: The cutoff frequency for the high-pass filter.

    Returns:
    - filtered_data: The high-pass filtered audio signal.
    """
    nyquist = 0.5 * sample_rate
    normal_cutoff = cutoff_freq / nyquist
    b, a = butter(1, normal_cutoff, btype='high', analog=False)
    filtered_data = filtfilt(b, a, data)
    return filtered_data

def process_wav_file(input_wav, output_wav, cutoff_freq=100):
    """
    Process a WAV file to remove DC component (frequencies below cutoff_freq).

    Parameters:
    - input_wav: Path to the input WAV file.
    - output_wav: Path to save the output WAV file.
    - cutoff_freq: The cutoff frequency for the high-pass filter.
    """
    # Read the input WAV file
    sample_rate, data = wavfile.read(input_wav)

    # Check if the data is stereo and convert to mono if necessary
    if len(data.shape) == 2:
        data = np.mean(data, axis=1)

    # Apply high-pass filter to remove DC component
    filtered_data = high_pass_filter(data, sample_rate, cutoff_freq)

    # Normalize the filtered data to avoid clipping
    filtered_data = np.int16(filtered_data / np.max(np.abs(filtered_data)) * 32767)

    # Write the filtered data to the output WAV file
    wavfile.write(output_wav, sample_rate, filtered_data)

# Example usage
input_wav = 'input.wav'   # Path to your input WAV file
output_wav = 'output.wav' # Path to save the processed WAV file
process_wav_file(input_wav, output_wav, cutoff_freq=100)
```
